```python
# Imprting required Packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
# Parameters for visualisation aesthetics
font = {'family' : 'Times',
        'weight' : 'bold',
        'size'   : 12}
matplotlib.rc('font', **font)
matplotlib.rcParams['figure.dpi'] = 600
matplotlib.rcParams["mathtext.fontset"]="dejavuserif"

# Input Path for Input Files Location
## Change this to Network 1 or Network 2 to produce figures S-4 and S-5 Respectively
path='../Network-Files/Network 3/'
network=path[-10:-3]+path[-2]

# Loading the results of 12 files (6 methods x 2 Supply durations)
FCVEM_4hr=pd.read_csv(path+network+'_4hr_FCV-EM_Means.csv',index_col=(0)).iloc[:,0]
FCVRes_4hr=pd.read_csv(path+network+'_4hr_FCV-Res_Means.csv',index_col=(0)).iloc[:,0]
CVTank_4hr=pd.read_csv(path+network+'_4hr_CV-Tank_Means.csv',index_col=(0)).iloc[:,0]
PSVTank_4hr=pd.read_csv(path+network+'_4hr_PSV-Tank_Means.csv',index_col=(0)).iloc[:,0]
PDA_4hr=pd.read_csv(path+network+'_4hr_PDA_Means.csv',index_col=(0)).iloc[:,0]
CVRes_4hr=pd.read_csv(path+network+'_4hr_CV-Res_Means.csv',index_col=(0)).iloc[:,0]
FCVEM_12hr=pd.read_csv(path+network+'_12hr_FCV-EM_Means.csv',index_col=(0)).iloc[:,0]
FCVRes_12hr=pd.read_csv(path+network+'_12hr_FCV-Res_Means.csv',index_col=(0)).iloc[:,0]
CVTank_12hr=pd.read_csv(path+network+'_12hr_CV-Tank_Means.csv',index_col=(0)).iloc[:,0]
PSVTank_12hr=pd.read_csv(path+network+'_12hr_PSV-Tank_Means.csv',index_col=(0)).iloc[:,0]
PDA_12hr=pd.read_csv(path+network+'_12hr_PDA_Means.csv',index_col=(0)).iloc[:,0]
CVRes_12hr=pd.read_csv(path+network+'_12hr_CV-Res_Means.csv',index_col=(0)).iloc[:,0]

# Plotting x axis 
xaxis=np.arange(0,4.00001,1/60)

# Line indicating the upper limit of flow-restriction
frestriction_x=[0,4]
frestriction_y=[0,100]

# 2 subplots (4 hours and 12 hours per day)
fig, axs=plt.subplots(1,2, gridspec_kw={'width_ratios': [1, 3]})
fig.set_figwidth(8)
fig.set_figheight(4)
#ax.set_title('Network 1')
axs[0].set_xlim(0, 4)
axs[0].set_ylim(0, 140)
axs[0].set_xticks(np.arange(0, 5, 4))
axs[0].set_xticks(np.arange(0, 5, 1), minor=True)
axs[0].set_yticks(np.arange(0, 141, 20))
axs[0].tick_params(axis='both', which='major', labelsize=12)
axs[0].tick_params(axis='both', which='minor', labelsize=12)

line3,= axs[0].plot(xaxis,PDA_4hr,color='#4575b4',label='EPANET-PDA',linewidth=2)
line4,= axs[0].plot(xaxis,FCVRes_4hr,color='#e0f3f8',label='FCV-Res',zorder=3,ls='dashed',linewidth=2)
line5,= axs[0].plot(xaxis,FCVEM_4hr,color='#91bfdb',label='FCV-EM',zorder=5, ls='dotted',linewidth=2)
line1, =axs[0].plot(xaxis,CVTank_4hr,color='#d73027',label='CV-Tank',zorder=4,linewidth=2)
line2, =axs[0].plot(xaxis,PSVTank_4hr,color='#fc8d59',label='PSV-Tank',ls='dotted',zorder=5,linewidth=2)
line6,=axs[0].plot(xaxis,CVRes_4hr,color='#fee090',label='CV-Res',linewidth=2)
line7,=axs[0].plot(frestriction_x, frestriction_y,color='black',ls='dashed')
# ax.annotate('$\mathrm{Q_{des}/V_{des}}$',(1.5,55),rotation=55,fontsize=10,fontweight="light")
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)
axs[0].set(ylabel='Satisfaction Ratio (%)')
axs[0].set(xlabel='Supply Time (hr)')



xaxis=np.arange(0,12.00001,1/60)

#ax.set_title('Network 1')
axs[1].set_xlim(0, 12)
axs[1].set_ylim(0, 140)
axs[1].set_xticks(np.arange(0, 13, 4))
axs[1].set_xticks(np.arange(0, 13, 1), minor=True)
axs[1].set_yticks(np.arange(0, 141, 20))
axs[1].tick_params(axis='both', which='major', labelsize=12)
axs[1].tick_params(axis='both', which='minor', labelsize=12)
line3,= axs[1].plot(xaxis,PDA_12hr,color='#4575b4',label='EPANET-PDA',linewidth=2)
line4,= axs[1].plot(xaxis,FCVRes_12hr,color='#e0f3f8',label='FCV-Res',zorder=3,ls='dashed',linewidth=2)
line5,= axs[1].plot(xaxis,FCVEM_12hr,color='#91bfdb',label='FCV-EM',zorder=5, ls='dotted',linewidth=2)
line1, =axs[1].plot(xaxis,CVTank_12hr,color='#d73027',label='CV-Tank',zorder=4,linewidth=2)
line2, =axs[1].plot(xaxis,PSVTank_12hr,color='#fc8d59',label='PSV-Tank',ls='dotted',zorder=5,linewidth=2)
line6,=axs[1].plot(xaxis,CVRes_12hr,color='#fee090',label='CV-Res',linewidth=2)
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
# plt.ylabel('S (%)')
axs[1].set(xlabel='Supply Time (hr)')
axs[1].legend(loc='lower right',fontsize='small',ncol=2)

# plt.savefig('/Users/omaraliamer/Desktop/UofT/Papers/Clean Network Files/Figure1_1.png',dpi=450)
plt.show()
```

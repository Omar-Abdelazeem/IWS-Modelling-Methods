```python
# Imprting required Packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib as mpl

# Parameters for visualisation aesthetics
mpl.rcParams['figure.dpi'] = 300
font = {'family' : 'Times',
        'weight' : 'bold',
        'size'   : 12}

mpl.rc('font', **font)

# Color scheme
fr_color='#4575b4'
swmmfr_color='#DE8D08'
vr_color='#d73027'
swmmvr_color='#383371'

# Input Path for Input Files Location
## Change this to Network 1 or Network 2 to produce figures S-9 and S-10 Respectively
path='../Network-Files/Network 3/'
network=path[-10:-3]+path[-2]

FR10_4hr=pd.read_csv(path+network+'_4hr_PDA_10thPercentile.csv')
FRmean_4hr=pd.read_csv(path+network+'_4hr_PDA_Means.csv')
FR90_4hr=pd.read_csv(path+network+'_4hr_PDA_90thPercentile.csv')

swmmFR10_4hr=pd.read_csv(path+network+'_4hr_Outlet-Outfall_10thPercentile.csv')
swmmFRmean_4hr=pd.read_csv(path+network+'_4hr_Outlet-Outfall_Means.csv')
swmmFR90_4hr=pd.read_csv(path+network+'_4hr_Outlet-Outfall_90thPercentile.csv')

FR10_12hr=pd.read_csv(path+network+'_12hr_PDA_10thPercentile.csv')
FRmean_12hr=pd.read_csv(path+network+'_12hr_PDA_Means.csv')
FR90_12hr=pd.read_csv(path+network+'_12hr_PDA_90thPercentile.csv')

swmmFR10_12hr=pd.read_csv(path+network+'_12hr_Outlet-Outfall_10thPercentile.csv')
swmmFRmean_12hr=pd.read_csv(path+network+'_12hr_Outlet-Outfall_Means.csv')
swmmFR90_12hr=pd.read_csv(path+network+'_12hr_Outlet-Outfall_90thPercentile.csv')

VR10_4hr=pd.read_csv(path+network+'_4hr_CV-Tank_10thPercentile.csv')
VRmean_4hr=pd.read_csv(path+network+'_4hr_CV-Tank_Means.csv')
VR90_4hr=pd.read_csv(path+network+'_4hr_CV-Tank_90thPercentile.csv')

swmmVR10_4hr=pd.read_csv(path+network+'_4hr_Outlet-Storage_10thPercentile.csv')
swmmVRmean_4hr=pd.read_csv(path+network+'_4hr_Outlet-Storage_Means.csv')
swmmVR90_4hr=pd.read_csv(path+network+'_4hr_Outlet-Storage_90thPercentile.csv')

VR10_12hr=pd.read_csv(path+network+'_12hr_CV-Tank_10thPercentile.csv')
VRmean_12hr=pd.read_csv(path+network+'_12hr_CV-Tank_Means.csv')
VR90_12hr=pd.read_csv(path+network+'_12hr_CV-Tank_90thPercentile.csv')

swmmVR10_12hr=pd.read_csv(path+network+'_12hr_Outlet-Storage_10thPercentile.csv')
swmmVRmean_12hr=pd.read_csv(path+network+'_12hr_Outlet-Storage_Means.csv')
swmmVR90_12hr=pd.read_csv(path+network+'_12hr_Outlet-Storage_90thPercentile.csv')

xs_4hr=np.arange(0,4.00001,1/60)
xsSWMM_4hr=swmmFR10_4hr.iloc[:,0]/3600

xs_12hr=np.arange(0,12.00001,1/60)
xsSWMM_12hr=swmmFR10_12hr.iloc[:,0]/3600


fig, axs=plt.subplots(2,2,gridspec_kw={'width_ratios': [1, 3]})
fig.set_figwidth(8)
fig.set_figheight(8)
axs[0,0].set_xlim(0, 4)
axs[0,0].set_ylim(0, 100)
axs[0,0].set_xticks(np.arange(0, 5, 1))
axs[0,0].set_xticks(np.arange(0, 5, 1), minor=True)
axs[0,0].set_yticks(np.arange(0, 101, 20))
axs[0,0].tick_params(axis='both', which='major', labelsize=12)
axs[0,0].tick_params(axis='both', which='minor', labelsize=12)
line1,=axs[0,0].plot(xs_4hr,FRmean_4hr.iloc[:,1],color=fr_color,label='Without Filling',zorder=5,linewidth=2)
line2,=axs[0,0].plot(xsSWMM_4hr,swmmFRmean_4hr.iloc[:,1],color=swmmfr_color,label='With Filling',linewidth=2,zorder=5)
axs[0,0].fill_between(xsSWMM_4hr, y1=swmmFR10_4hr.iloc[:,1], y2=swmmFR90_4hr.iloc[:,1], alpha=0.4,color=swmmfr_color,zorder=1,edgecolor=None,linewidth=0)
axs[0,0].fill_between(xs_4hr, y1=FR10_4hr.iloc[:,1], y2=FR90_4hr.iloc[:,1], alpha=0.4,color=fr_color,edgecolor=None,linewidth=0,zorder=4)
axs[0,0].spines['top'].set_visible(False)
axs[0,0].spines['right'].set_visible(False)
axs[0,0].set(ylabel='Satisfaction Ratio (%)')
axs[0,0].set(xlabel='Supply Time (hr)')

axs[0,1].set_xlim(0, 12)
axs[0,1].set_ylim(0, 101)
axs[0,1].set_xticks(np.arange(0, 13, 1))
axs[0,1].set_xticks(np.arange(0, 13, 1), minor=True)
axs[0,1].set_yticks(np.arange(0, 101, 20))
axs[0,1].tick_params(axis='both', which='major', labelsize=12)
axs[0,1].tick_params(axis='both', which='minor', labelsize=12)
line1,=axs[0,1].plot(xs_12hr,FRmean_12hr.iloc[:,1],color=fr_color,label='Steady-state',zorder=1,linewidth=2)
line2,=axs[0,1].plot(xsSWMM_12hr,swmmFRmean_12hr.iloc[:,1],color=swmmfr_color,label='Filling',linewidth=2)
axins=axs[0,1].inset_axes([0.1,0.8,0.2,0.2])
axins.set_xlim(0,1)
axins.set_ylim(0,10)
axins.plot(xs_12hr,FRmean_12hr.iloc[:,1],color=fr_color,label='FR',zorder=5,linewidth=2)
axins.plot(xsSWMM_12hr,swmmFRmean_12hr.iloc[:,1],color=swmmfr_color,label='SWMM-FR',linewidth=2)
axins.fill_between(xsSWMM_12hr, y1=swmmFR10_12hr.iloc[:,1], y2=swmmFR90_12hr.iloc[:,1], alpha=0.5,color=swmmfr_color,zorder=5,edgecolor=None,linewidth=0)
axins.fill_between(xs_12hr, y1=FR10_12hr.iloc[:,1], y2=FR90_12hr.iloc[:,1], alpha=0.5,color=fr_color,edgecolor=None,linewidth=0)
axins.set_yticks(np.arange(0,11,10),fontsize=20)
axins.set_xticks(np.arange(0,2,1),fontsize=0.5)
axins.tick_params(axis='both', labelsize=10)
axs[0,1].fill_between(xsSWMM_12hr, y1=swmmFR10_12hr.iloc[:,1], y2=swmmFR90_12hr.iloc[:,1], alpha=0.4,color=swmmfr_color,zorder=1,edgecolor=None,linewidth=0)
axs[0,1].fill_between(xs_12hr, y1=FR10_12hr.iloc[:,1], y2=FR90_12hr.iloc[:,1], alpha=0.4,color=fr_color,edgecolor=None,linewidth=0)
axs[0,1].spines['top'].set_visible(False)
axs[0,1].spines['right'].set_visible(False)
axs[0,1].set(xlabel='Supply Time (hr)')
axs[0,1].legend(loc='lower right',fontsize='small',ncol=1)

axs[1,0].set_xlim(0, 4)
axs[1,0].set_ylim(0, 100)
axs[1,0].set_xticks(np.arange(0, 5, 1))
axs[1,0].set_xticks(np.arange(0, 5, 1), minor=True)
axs[1,0].set_yticks(np.arange(0, 101, 20))
axs[1,0].tick_params(axis='both', which='major', labelsize=12)
axs[1,0].tick_params(axis='both', which='minor', labelsize=12)
line1,=axs[1,0].plot(xs_4hr,VRmean_4hr.iloc[:,1],color=vr_color,label='Without Filling',zorder=1,linewidth=2)
line2,=axs[1,0].plot(xsSWMM_4hr,swmmVRmean_4hr.iloc[:,1],color=swmmvr_color,label='With Filling',linewidth=2)
axs[1,0].fill_between(xsSWMM_4hr, y1=swmmVR10_4hr.iloc[:,1], y2=swmmVR90_4hr.iloc[:,1], alpha=0.4,color=swmmvr_color,zorder=1,edgecolor=None,linewidth=0)
axs[1,0].fill_between(xs_4hr, y1=VR10_4hr.iloc[:,1], y2=VR90_4hr.iloc[:,1], alpha=0.4,color=vr_color,edgecolor=None,linewidth=0)
axs[1,0].spines['top'].set_visible(False)
axs[1,0].spines['right'].set_visible(False)
axs[1,0].set(ylabel='Satisfaction Ratio (%)')
axs[1,0].set(xlabel='Supply Time (hr)')

axs[1,1].set_xlim(0, 12)
axs[1,1].set_ylim(0, 101)
axs[1,1].set_xticks(np.arange(0, 13, 1))
axs[1,1].set_xticks(np.arange(0, 13, 1), minor=True)
axs[1,1].set_yticks(np.arange(0, 101, 20))
axs[1,1].tick_params(axis='both', which='major', labelsize=12)
axs[1,1].tick_params(axis='both', which='minor', labelsize=12)
# line3,=plt.plot(fakelinesx,fakelinesy,color=fr_color,label='FR',zorder=5,linewidth=2)
# line4,=plt.plot(fakelinesx,fakelinesy,color=swmmfr_color,label='SWMM-FR',linewidth=2)
line1,=plt.plot(xs_12hr,VRmean_12hr.iloc[:,1],color=vr_color,label='Steady-State',zorder=1,linewidth=2)
line2,=plt.plot(xsSWMM_12hr,swmmVRmean_12hr.iloc[:,1],color=swmmvr_color,label='Filling',linewidth=2)
axins=axs[1,1].inset_axes([0.1,0.8,0.2,0.2])
axins.set_xlim(0,1)
axins.set_ylim(0,10)
axins.plot(xs_12hr,VRmean_12hr.iloc[:,1],color=vr_color,label='Steady-State',zorder=5,linewidth=2)
axins.plot(xsSWMM_12hr,swmmVRmean_12hr.iloc[:,1],color=swmmvr_color,label='Filling',linewidth=2)
axins.fill_between(xsSWMM_12hr, y1=swmmVR10_12hr.iloc[:,1], y2=swmmVR90_12hr.iloc[:,1], alpha=0.5,color=swmmvr_color,zorder=5,edgecolor=None,linewidth=0)
axins.fill_between(xs_12hr, y1=VR10_12hr.iloc[:,1], y2=VR90_12hr.iloc[:,1], alpha=0.5,color=vr_color,edgecolor=None,linewidth=0)
axins.set_yticks(np.arange(0,11,10),fontsize=1)
axins.set_xticks(np.arange(0,2,1),fontsize=0.5)
axins.tick_params(axis='both', labelsize=10)
axs[1,1].fill_between(xsSWMM_12hr, y1=swmmVR10_12hr.iloc[:,1], y2=swmmVR90_12hr.iloc[:,1], alpha=0.4,color=swmmvr_color,zorder=1,edgecolor=None,linewidth=0)
axs[1,1].fill_between(xs_12hr, y1=VR10_12hr.iloc[:,1], y2=VR90_12hr.iloc[:,1], alpha=0.4,color=vr_color,edgecolor=None,linewidth=0)
axs[1,1].spines['top'].set_visible(False)
axs[1,1].spines['right'].set_visible(False)
# plt.ylabel('Satisfaction Ratio (%)')
axs[1,1].set(xlabel='Supply Time (hr)')
axs[1,1].legend(loc='lower right',fontsize='small',ncol=1)

plt.show()
```
